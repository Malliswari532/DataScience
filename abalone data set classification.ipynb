# Auto ML


```python
from sklearn.ensemble import AdaBoostClassifier,GradientBoostingClassifier,RandomForestClassifier
```

```python
from sklearn.svm import SVC
```

```python
from xgboost import XGBRFClassifier
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
adc=AdaBoostClassifier()
```

```python
gbc=GradientBoostingClassifier()
```

```python
rfc=RandomForestClassifier()
```

```python
svm=SVC()
```

```python
xgb=XGBRFClassifier()
```

```python
dtc=DecisionTreeClassifier()
```

```python
clist=[adc,gbc,rfc,svm,xgb,dtc]
```

```python
clist
```

```python
import time
```

```python
for i in clist:
    print(i,end = "==")
    i.fit(X_train,y_train)
    m=i.score(X_test,y_test)
    print(m)
    time.sleep(0.6)
```

# K fold cross validation

```python
from sklearn.model_selection import KFold
```

```python
kfold_validation=KFold(10)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
clist
```

```python
for mod in clist:
    print(mod,end="==>")
    score=cross_val_score(mod,X,y,cv=kfold_validation)
    print(score)
    print("Average of scores",np.mean(score))
    print("**************************************************************")
```

# stratified kfold cross validation

```python
from sklearn.model_selection import StratifiedKFold
```

```python
skfold_validation=StratifiedKFold(10)
```

```python
for mod in clist:
    print(mod,end="==>")
    scores=cross_val_score(mod,X,y,cv=skfold_validation)
    print(scores)
    print("Average of scores",np.mean(scores))
    print("*********************************************************")
```

# Leave one out cross validation

```python
from sklearn.model_selection import LeaveOneOut
```

```python
Loocv=LeaveOneOut()
```

```python
for mod in clist:
    print(mod,end="==>")
    s=cross_val_score(mod,X,y,cv=Loocv)
    print(s)
    print("Average of scores",np.mean(s))
    print("*********************************************************")
```
